Monte Carlo Pg 6

```python
# monte_carlo_estimate.py
import random
import numpy as np

def uniform_random(n, seed=None):
    """Generate n uniform random numbers in [0,1]."""
    if seed is not None:
        random.seed(seed)
    return [random.random() for _ in range(n)]

def f(x):
    """Example function f(x) = x^2"""
    return x**2

if __name__ == "__main__":
    # Step 1: generate 10 uniform samples
    samples = uniform_random(10, seed=42)

    # Step 2: evaluate f for each sample
    f_values = [f(x) for x in samples]

    # Print values (rounded similar to screenshot)
    print("Random Samples (x):", np.round(samples, 4))
    print("Function Values f(x):", np.round(f_values, 4))

    # Step 3: Monte Carlo estimate of integral (mean of f_values over [0,1])
    I_est = sum(f_values) / len(f_values)
    print("\nMonte Carlo Integral Estimate I =", round(I_est, 4))
```

```python
# kmeans_clustering_iris.py
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd

if __name__ == "__main__":
    # Load unsupervised dataset (ignore labels)
    iris = load_iris()
    X = iris.data  # only features

    # Apply K-Means clustering
    kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)

    # Show results in a DataFrame
    df = pd.DataFrame(X, columns=['SepalLen', 'SepalWid', 'PetalLen', 'PetalWid'])
    df['Cluster'] = labels
    print(df.head())
```

```python

!pip install pgmpy
# bayesian_network_pgmpy.py
import pandas as pd
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination

if __name__ == "__main__":
    # Step 1: Create a small realistic dataset
    data = pd.DataFrame({
        'Rain':    ['Yes','Yes','No','No','Yes','No','Yes','No','No','Yes'],
        'Traffic': ['High','High','Low','Low','Low','Low','High','Low','High','High'],
        'Late':    ['Yes','No','No','No','Yes','No','Yes','No','No','Yes']
    })

    # Step 2: Define Bayesian Network structure (Rain -> Traffic -> Late)
    model = DiscreteBayesianNetwork([('Rain', 'Traffic'), ('Traffic', 'Late')])

    # Step 3: Train (estimate CPDs) using Maximum Likelihood
    model.fit(data, estimator=MaximumLikelihoodEstimator)

    # Step 4: Inference
    inference = VariableElimination(model)

    # Step 5: Query P(Late | Rain = 'Yes')
    prob = inference.query(variables=['Late'], evidence={'Rain': 'Yes'})
    print(prob)
```

```python
# If you're in a Jupyter notebook, run this cell first to install:
!pip install minisom

from minisom import MiniSom
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Load dataset (unsupervised: only features)
X = load_iris().data

# Normalize data
X = MinMaxScaler().fit_transform(X)

# Initialize and train SOM
som = MiniSom(x=5, y=5, input_len=X.shape[1], sigma=1.0, learning_rate=0.5)
som.random_weights_init(X)
som.train_random(X, num_iteration=100)

# Map data points to clusters (winning neuron coordinates)
win_map = np.array([som.winner(x) for x in X])
print("Sample SOM neuron mappings (first 10):\n", win_map[:10])
```
